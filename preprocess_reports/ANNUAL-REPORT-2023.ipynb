```python
# Generate prompts: 360x7=2520 prompts
data = parsed_doc.model_dump()

# Create a dictionary to hold the context strings
context_strs = {}
for n in range(360):
    context_strs[f'context_str_{n+1}'] = data['nodes'][n]['text']

# Combine each context with each template
combined_texts = []
for key, context in context_strs.items():
    for template_name, template in templates.items():
        combined_text = template.format(context_str=context)
        combined_texts.append((template_name, key, combined_text))

# combined_texts[0][(template_name, context_key, combined_text)]
# print(combined_texts[0][2])
```

#### randomly select 100(150) promtps to annotate

```python
import random
import csv

# Select 20 random prompts from the list
random_prompts = random.sample(combined_texts, 150)

# Print the selected prompts
for i, prompt in enumerate(random_prompts, start=1):
    template_name, context_key, combined_text = prompt
    print(f"Prompt {i} ({template_name}, {context_key}): {combined_text}")

# Save the selected prompts to a CSV file
csv_file_path = 'random_prompts.csv'
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write the header
    writer.writerow(['Prompt Number', 'Template Name', 'Context Key', 'Combined Text'])
    # Write the selected prompts
    for i, prompt in enumerate(random_prompts, start=1):
        template_name, context_key, combined_text = prompt
        writer.writerow([i, template_name, context_key, combined_text])

print(f"{len(random_prompts)} prompts have been saved to {csv_file_path}")
```

#### Generate all triples with llama3, vicuna, mistral-inst

```python
from llama_index.llms.ollama import Ollama
import time

llm = Ollama(model="llama3")

try:
    for i in range(len(combined_texts)):
        prompt = combined_texts[i][2]
        print(f"Processing prompt {i + 1} out of {len(combined_texts)}")
        start_time = time.time()
        response = llm.complete(prompt) 
        end_time = time.time()
        print(f"Query LLM processing time: {end_time - start_time}")
        print(f"Model's Answer for prompt {i + 1}:\n{response}\n")
except Exception as e:
    print(f"An error occurred: {e}")
```
